	<p> $C$  $c_1c_2 ... c_k$$k$ $p(c_1)p(c_2) ... p(c_k)$ Entropy :</p>
	<p>$$info(C) = - \sum_{i=1}^k p(c_i) \cdot log_2 p(c_i) \qquad (fml.4.1.1)$$</p>
	<p>ID  Category  $c_1c_2 ... c_k$  (fml.4.1.1) </p>

<p><strong>Condition Entropy</strong></p>
<p>  $f$  $I$ $info(C|f=f_i)$  $f$  $f_i$ <strong><em>Condition Entropy</em></strong> $info(C|f)$  $f$  <strong><em>Condition Entropy</em></strong></p>

<blockquote>
	$$\begin{align} info(C|f) & = p_1 \cdot info(C|f=f_1) + p_2 \cdot info(C|f=f_2) + ... + p_k \cdot info(C|f=f_{k}) \\ & = \sum_{i=1}^{I} p_i \cdot info(C|f=f_i) \end{align}  \quad (fml.4.1.2)$$
    $m$  $f=f_i$  $m_ip_i=\frac{m_i}{m}$.
</blockquote>

   :</p>
	<p> $$\begin{align} info(C|f=f_i) = -\sum_{j=0}^{k-1} \frac{m_{ij}}{m_i} \cdot log_{2} \frac{m_{ij}}{m_i} \end{align} \qquad (fml.4.1.4)$$</p>
	<p>$\frac{m_{ij}}{m_i}$ $f=f_i$  $C=c_j$  $p(c_j|f_i)$</p>
