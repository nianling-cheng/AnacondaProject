### 4.1.3 数值型数据: 平均数(Mean) <br/>
#### 1 简单平均数与加权平均数 <br/>
##### $简单平均数 \ \overline{x}=\frac{x_1+x_2+...+x_n}{n}=\frac{\sum_{i=1}^{n}{x_i}}{n}$ <br/>
##### $加权平均数 \ \overline{x}=\frac{M_1f_1+M_2f_2+...+M_nf_n}{f_1+f_2+...+f_n}=\frac{\sum_{i=1}^{n}{M_if_i}}{N} \ , \ 公式中\ N=\sum{f_i} \ ,\ 即为样本容量$ <br/>
#### 2 平均数的另一种表现形式: 调和平均数(Harmonie Mean) <br/>
#### 3 一种特殊的平均数: 几何平均数(Geometric Mean): $\large{G_m=\sqrt[n]{x_1x_2...x_n}=\sqrt[n]{\prod_{i=1}^{n}{x_i}}}$ <br/>
设开始的数值为 $y_0$ , 逐年增长率为 $G_1 , G_2 , ... , G_n$ , 第 $n$ 年的数值为 $y_n=y_0 (1+G_1) (1+G_2) ... (1+G_n)=y_0 \prod_{i=1}^{n}{(1+G_i)}$ . 从 $y_0$ 到 $y_n$ 用 $n$ 年, 每年的增长率都相同, 这个增长率 $G$ 就是平均增长率 $\overline{G}$ , 即公式中的 $G_i$ 都等于 $G$ . 因此 $(1+G)^n=\prod_{i=1}^{n}{(1+G_i)}$ , 则 $\overline{G}=\sqrt[n]{\prod_{i=1}^{n}{(1+G_i)} } -1$  <br/>

## 4.2 离散程度度量 <br/>
### 4.2.1 分类数据: 异众比率(Variation Ratio) $V_r=\frac{\sum{f_i} - f_m}{\sum{f_i}}=1-\frac{f_m}{\sum{f_i}} \ , \ \sum{f_i}表示变量值的总频数, \ {f_m} \ 表示众数组的频数$ <br/>
### 4.2.2 顺序数据: 四分位差(Quartile Deviation, Inter-Quartile Range) $Q_d=Q_U-Q_L$ <br/>
### 4.2.3 数值型数据: 方差(Variance)和标准差(Sandard Deviation) <br/>
#### 1 极差(Range): $R=max(x_i)-min(x_i)$ <br/>
#### 2 平均差(平均离差, Mean Deviation) <br/>
未分组数据平均差: $\large{M_d=\frac{\sum_{i=1}^{n}{| x_i - \overline{x} |}}{n}}$ <br/>
分组数据平均差: $\large{M_d=\frac{\sum_{i=1}^{k}{| M_i - \overline{x} | f_i}}{n}}$ <br/>
#### 3 方差(Variance)和标准差(Standard Deviation) <br/>
未分组数据: $\large{s^2=\frac{\sum_{i=1}^{n}{(x_i - \overline{x})^2}}{n-1}}$ <br/>
分组数据: $\large{s^2=\frac{\sum_{i=1}^{k}{(M_i - \overline{x})^2 f_i}}{n-1}}$ <br/>
样本方差就是用样本数据个数或总频数减 1 去除离差平方和, 其中样本数据个数减 1 即 n-1 称为自由度(Degree of Freedom) <br/>
未分组数据: $\large{s=\sqrt{\frac{\sum_{i=1}^{n}{(x_i - \overline{x})^2}}{n-1}}}$ <br/>
分组数据: $\large{s=\sqrt{\frac{\sum_{i=1}^{k}{(M_i - \overline{x})^2 f_i}}{n-1}}}$ <br/>
### 4.2.4 相对位置的度量: 标准分数 <br/>
#### 1 标准分数(Standard Score) $\large{z_i=\frac{x_i - \overline{x}}{s}}$ <br/>
标准分数具有平均数为 0 , 标准差为 1 的特性 <br/>
#### 2 经验法则 <br/>
约有 68% 的数据在平均数加减 1 个标准差范围内 <br/>
约有 95% 的数据在平均数加减 2 个标准差范围内 <br/>
约有 99% 的数据在平均数加减 3 个标准差范围内 <br/>
在 3 个标准差范围外的数据称为异常值或离群点(Outlier) <br/>
#### 3 切比雪夫不等式 <br/>
对于任意分布形态的数据, 至少有 $(1 - 1/k^2)$ 的数据在 k 个标准差之内. 其中 k 是大于 1 的任意值, 不一定是整数.  <br/>
### 4.2.5 相对离散程度: 离散系数(变异系数, Coefficient of Variation) $\large{v_s=\frac{s}{\overline{x}}}$ <br/>
分类数据: 异众比率 <br/>
顺序数据: 四分位差 <br/>
数值型数据: 方差, 标准差 <br/>
## 4.3 偏态与峰态的度量 <br/>
### 4.3.1 偏态(Skewness)及其测度 <br/>
未分组数据偏态系数: $\large{SK=\frac{n\sum{(x_i - \overline{x})^3}}{(n-1)(n-2)s^3}}$ <br/>
分组数据偏态系数: $\large{SK=\frac{n\sum_{i=1}^{k}{(M_i - \overline{x})^3 f_i}}{ns^3}}$ <br/>
### 4.3.2 峰态(Kurtosis)及其测度 <br/>
未分组数据峰态系数: $\large{K=\frac{n(n+1) \sum{(x_i - \overline{x})^4 - 3 [\sum{(x_i - \overline{x})^2}]^2 (n-1) } }{(n-1)(n-2)(n-3) s^4}}$ <br/>
分组数据峰态系数: $\large{K=\frac{\sum_{i=1}^{k}{(M_i - \overline{x})^4 f_i} }{ns^4} -3}$ <br/>

# 第五章 概率与概率分布
## 5.1 事件及其概率
### 5.1.1 实验(Experiment), 事件(Event)和样本空间(Sample Space, $\omg$)
简单事件(Simple Event); 必然事件 $\omg$ ; 不可能事件 $\none$; 样本点(Sample Point, $\omg$)
### 5.1.2 事件的概率(Probaility) $\large{P(A)=\frac{事件A发生的次数}{重复试验次数}=\frac{m}{n}=p}$
### 5.1.3 概率的性质和运算法则
#### 1 互斥事件(Mutually Exclusive Events)及其概率
文氏图(Venn Diagram)
互斥事件加法规则: $P(AUB)=P(A)+P(B)$
#### 2 事件的补(Complement , $\overline{A}$)及其概率: $P(\overline{A})=1-P(A)$
#### 3 广义加法公式: $P(AUB)=P(A)+P(B)-P(AB)$
事件的并(Union): $AUB$
事件的交(Intersection): $AB$
### 5.1.4 条件概率与事件的独立性
#### 1 条件概率(Conditional Probability, $P(A|B)$)
条件概率计算公式: $\large{P(A|B)=\frac{P(AB)}{P(B)}}$ . P(AB)称为联合概率, P(A), P(B)均称为边际概率
#### 2 乘法公式: $P(AB)=P(B) P(A|B)=P(A) P(B|A)$
#### 3 独立事件(Independent Events): 如果 $P(A|B)=P(A)$ 或 $P(B|A)=P(B)$
独立事件乘法公式: $P(AB)=P(A) P(B)$ 或 $P(A_1 , A_2 , ... , A_n)=P(A_1)P(A_2)...P(A_n)$
### 5.1.5 全概率公式与逆概率公式
#### 1 全概率公式: ${P(A)=\sum_{i=1}^{n}{P(AB_i)}=\sum_{i=1}^{n}{P(B_i)P(A|B_i)} , 其中 B_1 , B_2 , ... , B_n 是互不相容的事件且 B_1 U B_2 U ... U B_n=\omg}$
#### 2 逆概率公式: $\large{P(B_j |A)}=\frac{P(B_j) P(A|B_j)}{\sum_{i=1}^{n}{P(B_i) P(A|B_i)}} , 其中 B_1 , B_2 , ... , B_n 为完备事件组$
